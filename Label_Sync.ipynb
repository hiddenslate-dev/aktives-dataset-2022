```python
# NaN values may occur due to seconds difference

from functools import reduce

data_frames = [expert1,expert2,expert3]

df = reduce(lambda  left,right: pd.merge(left,right,on=['Second'],
                                            how='outer'), data_frames)
```

```python
df
```

```python
df = df.drop(columns = ["Minute_x","Minute_y","Minute"])
df.dropna(inplace = True)
```

```python
dt = df.copy()
```

```python
# Label Sync for "Stresli / Stressiz"

df['Stress/No Stress'] = df.apply(lambda x: x['Stress/No Stress-1'] if x['Stress/No Stress-1'] ==
                     x['Stress/No Stress-2'] or x['Stress/No Stress-1']
                     == x['Stress/No Stress-3'] else x["Stress/No Stress-2"], axis=1)
```

```python
# Label Sync for "Tepki Var / Tepki Yok"

df['Reaction/No Reaction'] = df.apply(lambda x: x['Reaction/No Reaction-1'] if x['Reaction/No Reaction-1'] ==
                     x['Reaction/No Reaction-2'] or x['Reaction/No Reaction-1']
                     == x['Reaction/No Reaction-3'] else x["Reaction/No Reaction-2"], axis=1)
```
